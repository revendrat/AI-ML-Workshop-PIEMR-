# **Box Plot**

* Box & Whisker plot is used to understand the variability and existence of outliers in data.

* Box plot is constructed using following elements:
 * First Quartile, Second Quartile (Median) & Third Quartile
 * Lowest & highest values
 * Inter Quartile Range (IQR)

* IQR is the distance between 3rd Quartile and 1st Quartile.

* Whisker of Box plot extends till (Q1 - 1.5*IQR) and (Q3 + 1.5*IQR).

* Observations beyond above limits are potential outliers.


```python
box = plt.boxplot(ipl_auction['sold_price'])
```

```python
# Finding Outliers
# 'caps' key in box variable returns the min and max values of the distribution 
[item.get_ydata()[0] for item in box['caps']]
```

```python
# 'whisker' key in box variable returns the values of distribution at 25 and 75 quartile
[item.get_ydata()[0] for item in box['whiskers']] 
```

```python
# 'median' key returns the median values of the distribution
[item.get_ydata()[0] for item in box['medians']]
```

```python
# Who are the Outliers?
ipl_auction[ipl_auction['sold_price'] > 1350000.0][['player_name', 'role', 'sold_price']] 
```
